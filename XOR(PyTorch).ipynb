```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
x_train = torch.FloatTensor([[0,0], [0,1], [1,0], [1,1]])
y_train = torch.FloatTensor([[0], [1], [1], [0]])

class class_BinaryClassification(nn.Module):
  def __init__(self):
    super().__init__()
    self.HL1=nn.Linear(2,4)
    self.Out=nn.Linear(4,1)
    self.sigmoid=nn.Sigmoid()
  def forward(self,x):
    HL1=self.sigmoid(self.HL1(x))
    return self.sigmoid(self.Out(HL1))

model_BinaryClassification = class_BinaryClassification()
optimizer = torch.optim.SGD(model_BinaryClassification.parameters(), lr=0.1)
	
for step in range(20000):
	prediction = model_BinaryClassification(x_train)
	cost = F.binary_cross_entropy(prediction, y_train)
	optimizer.zero_grad()
	cost.backward()
	optimizer.step()

model_test = model_BinaryClassification(x_train)
print(model_test)
model_test_binary = np.round(model_test>0.5).type(torch.float32)
print(model_test_binary)
```
